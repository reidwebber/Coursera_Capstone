```python
!conda install -c anaconda beautifulsoup4

import pandas as pd
import requests
from bs4 import BeautifulSoup
req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(req.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))
neighborhood=pd.DataFrame(df[0])
```

```python
neighborhood
```

```python
print(neighborhood['Postcode'].unique())
print(neighborhood['Borough'].unique())
print(neighborhood['Neighbourhood'].unique())
```

```python
# Making a list of missing value types
missing_values = ["Not assigned"]

# Looking at the Borough column
print (neighborhood ['Borough'])
print (neighborhood ['Borough'].isnull())
```

```python
# Delete the rows with label "Not assigned"
# For label-based deletion, set the index first on the dataframe:
neighborhood = neighborhood.set_index("Borough")
neighborhood = neighborhood.drop("Not assigned", axis=0)                              
```

```python
neighborhood
```

```python
# Combine the Neighborhoods for each Postcode
neighborhood = neighborhood.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
```

```python
#Get the geospatial data
geo = pd.read_csv("http://cocl.us/Geospatial_data/Geospatial_Coordinates.csv")

# take a look at the dataset
geo.head()
```

```python
#Rename the column "Postal Code" to match the Neighborhood column
geo = geo.rename(columns={'Postal Code': 'Postcode'})
geo.head(5)
```

```python
#Merge the two dataframes to create one set of data
toronto = pd.merge(neighborhood,geo, on='Postcode')
toronto.head()
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {43.74}, {-79.37}.')
```

```python
#create the map of Toronto
map_toronto = folium.Map(location=[43.74, -79.37], zoom_start=10)

#add markers to the map
for lat, lng, postcode, borough, neighborhood in zip(toronto['Latitude'], toronto['Longitude'], toronto['Postcode'],toronto['Borough'], toronto['Neighbourhood']):
    label = '{}, {}, {}'.format(postcode, borough, neighborhood)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  

map_toronto
```
